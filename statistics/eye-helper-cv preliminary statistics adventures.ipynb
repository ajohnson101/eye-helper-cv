The goal of this ipython notebook is to document our first attempts at statistical analysis for comparing object tracking methods in eye-helper-cv. ### Methods we're comparing
- SIFT
- SURF
- ORB
- BRISK

```python
#Code for our preliminary statistics adventures

import cv2
import numpy as np 
import csv
import pickle
import scipy.ndimage 
import scipy.stats
import os
import math
import matplotlib.pyplot as plt  
import pprint as pp
from IPython.display import display, Image


dstr = 'cookie' #choose your dataset.
framerange = [124, 288]
methods = ['SIFT', 'SURF', 'ORB', 'BRISK']

accuracy = False #default False 
distance = False #default False

#holds information we need
d_from_c = {} 
diffs = {} #diffs is short for differences in accuracy between SIFT and methods per trial

for trial in range(framerange[0], framerange[1], 20):
    d_from_c[trial] = {}
    for mstr in methods:
        #Loads the pickles which have all the possible data that we could want for the d_from_c work and puts them in d_from_c
        trialdata = pickle.load(open('../OT-res/compare_kpd_plots/%s_%s.p' % (dstr, mstr), 'rb'))[trial]
        d_from_c[trial][mstr] = [trialdata['distance from center'][x]/trialdata['hypotenuse'][x] for x in range(len(trialdata['distance from center']))]

#Must make a new loop because the fully created d_from_c is required for these things
for trial in range(framerange[0], framerange[1], 20):
    diffs[trial] = {}
    try:
        for mstr in methods:
            if mstr != 'SIFT':
                # Calculating the differences in accuracy between SIFT and methods per trial
                diffs[trial][mstr] = sum([float(d_from_c[trial][mstr][x] - d_from_c[trial]['SIFT'][x])/d_from_c[trial]['SIFT'][x] for x in range(len(d_from_c[trial][mstr]))])/len(d_from_c[trial][mstr])*100
        display(Image(filename='../OT-res/cookie_dfc_plots/cookie_trial%d_dfc_plot_sigma_two.png' % trial, retina=True)) 
        print "Trial %d" % trial
        pp.pprint(diffs[trial])
    except:
        # print 'yo'
        continue
```
